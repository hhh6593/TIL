```python
#employee 칼럼 전처리
a=test['employee1'].astype(str).tolist()
a2=test['employee2'].astype(str).tolist()
b=[]
b2=[]
for i in a:
    if ',' in i:
        i=i.replace(',','')
        b.append(i)
    else:
        b.append(i)
for i in a2:
    if ',' in i:
        i=i.replace(',','')
        b2.append(i)
    else:
        b2.append(i)
b=list(map(float, b))
b2=list(map(float, b2))
test['employee1']=b
test['employee2']=b2
```

```python
alldata2=alldata.drop(['inst_id','OC','sgg','openDate'], axis=1)
alldata2
```

```python
le = LabelEncoder()
le_col = ['sido','instkind','ownerChange']
for i in le_col:
    alldata2[i]=le.fit_transform(list(alldata2[i]))
alldata2
```

```python
alldata3=alldata2.fillna(-1)
alldata3
```

```python
train2=alldata3[:len(train)]
test2=alldata3[len(train):]
```

```python
alldata3.isnull().sum()
```

```python
train['OC']=le.fit_transform(train['OC'])
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
params = { 'n_estimators' : [3 ,7, 10, 15, 20, 50, 100] }
grid_cv = GridSearchCV(rf, param_grid=params, n_jobs=-1, cv=5)
grid_cv.fit(train2, train['OC'])

print('best parameter :', grid_cv.best_params_)
print('best score :', grid_cv.best_score_)
```

```python
grid_cv.best_estimator_
```

```python
rf=RandomForestClassifier(n_jobs=-1, 
                          random_state=42,
                          oob_score=True) #oob_score = 모델 자체적 교차검증 실시
rf.fit(train2, train['OC'])
result=rf.predict(test2)
```

```python
rf.oob_score_
```

```python
a=pd.Series(rf.feature_importances_, index=train2.columns).sort_values(ascending=False)
a
```

```python
#가장 영항을 많이 준 칼럼 상위 6개 분리
top_6 = a[:6].index
top_6
```

```python
#top_6만 사용해서 모델링
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score, cross_validate
scaler = MinMaxScaler(feature_range=(0,1)) #모든 값을 0~1사이로 표준화 
scaled_data = scaler.fit_transform(train2[top_6])
rf.fit(scaled_data, train['OC'])
predict = rf.predict(test2[top_6])
```

```python
from sklearn.model_selection import StratifiedKFold
sk = StratifiedKFold(n_splits=5, random_state=42, shuffle=True)
cr=cross_val_score(rf, train2[top_6], train['OC'], cv=sk)
cr
```
