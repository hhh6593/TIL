```python
list=[7,1,9,5,3,8,4,6,4,2]
for i in range(len(list)-1):
    min_idx=i
    for c in range(i+1, len(list)):
        if list[c]<list[min_idx]:
            min_idx=c
        list[i], list[min_idx]=list[min_idx], list[i]
        print(list)
print('출력물------------------------')
print(list)
```

```python
in_list=[5,1,6,9,2,7,0,3,8,4]
for i in range(1, len(in_list)):
    key=in_list[i]
    j=i-1
    while j>=0 and in_list[j]>key:
        in_list[j+1]=in_list[j]
        j=j-1
    in_list[j+1]=key
    print(in_list)
print('출력물------------------------')
print(in_list)
```
